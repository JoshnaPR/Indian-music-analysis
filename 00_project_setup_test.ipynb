```python
# Project Setup Test
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

print(" Indian Music Analysis Project")
print("=" * 40)
print("Testing all packages...")

# Test pandas
df = pd.DataFrame({
    'language': ['Hindi', 'Tamil', 'Telugu', 'Punjabi', 'Bengali'],
    'popularity': [85, 72, 68, 79, 65],
    'regions': ['North/West', 'South', 'South', 'North', 'East']
})

print(f" Pandas working - Created sample data with {len(df)} languages")

# Test visualization
plt.figure(figsize=(10, 6))
plt.bar(df['language'], df['popularity'], color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
plt.title('Sample: Indian Language Music Popularity', fontsize=16, fontweight='bold')
plt.xlabel('Language')
plt.ylabel('Popularity Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(" Matplotlib working - Chart created")
print(" All packages loaded successfully!")
print("\n Ready to analyze Indian music data!")
```
